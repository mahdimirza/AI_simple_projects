```python
####implementation of Pac-man#######################################

import random 
import copy
states = list()
statesIndx = list()

class Agent:
    def __init__(self, gama, alpha):
        self.currentState = [0, 0]
        self.gama = gama
        self.alpha = alpha

###################################################################################################
    def Terminate(self, env):
        numDots = 0
        for i in range(len(env)):
            for j in range(len(env[i])):
                if(env[i][j] == 'D'):
                    numDots += 1

        return(numDots)
###################################################################################################
    def UpadteQvalue(self, envprev, envnext, action, reward):
        if(envnext not in states):
            states.append(envnext)
            statesIndx.append([0, 0, 0, 0])
        
        action = int(action)

        indxprev = states.index(envprev)
        indxnext = states.index(envnext)

        Lindx1 = statesIndx[indxprev]
        Lindx2 = statesIndx[indxnext]

        Lindx1[action] = Lindx1[action] * (1 - self.alpha) + self.alpha * (reward + self.gama * (max(Lindx2)))
        statesIndx[indxprev] = Lindx1
###################################################################################################
    def UpAction(self,env):

        crdnt = self.GetCurrentState(env)
        env1 = copy.deepcopy(env)
        if(crdnt[0] >= 1):
            if(env1[crdnt[0] - 1][crdnt[1]] == 'D'):
               env1[crdnt[0] - 1][crdnt[1]] = 'A'
               env1[crdnt[0]][crdnt[1]] = 'E'
               self.currentState = [crdnt[0] - 1, crdnt[1]]
               return(env1, 1)
            if(env1[crdnt[0] - 1][crdnt[1]] == 'W'):
                return(env1, 0)
            if(env1[crdnt[0] - 1][crdnt[1]] == 'E'):
               env1[crdnt[0] - 1][crdnt[1]] = 'A'
               env1[crdnt[0]][crdnt[1]] = 'E'
               self.currentState = [crdnt[0] - 1, crdnt[1]]
               return(env1, 0)
            if(env1[crdnt[0] - 1][crdnt[1]] == 'G'):
               env1[crdnt[0] - 1][crdnt[1]] = 'A'
               env1[crdnt[0]][crdnt[1]] = 'E'
               self.currentState = [crdnt[0] - 1, crdnt[1]]
               return(env1, -100)
        else:
            return(env1, 0)
###################################################################################################
###################################################################################################
    def RightAction(self, env):

        crdnt = self.GetCurrentState(env)
        env1 = copy.deepcopy(env)
        if(crdnt[1] <= len(env[0]) - 2):
            if(env1[crdnt[0]][crdnt[1] + 1] == 'D'):
               env1[crdnt[0]][crdnt[1] + 1] = 'A'
               env1[crdnt[0]][crdnt[1]] = 'E'
               self.currentState = [crdnt[0], crdnt[1] + 1]
               return(env1, 1)
            if(env1[crdnt[0]][crdnt[1] + 1] == 'W'):
                return(env1, 0)
            if(env1[crdnt[0]][crdnt[1] + 1] == 'E'):
               env1[crdnt[0]][crdnt[1] + 1] = 'A'
               env1[crdnt[0]][crdnt[1]] = 'E'
               self.currentState = [crdnt[0], crdnt[1] + 1]
               return(env1, 0)
            if(env1[crdnt[0] - 1][crdnt[1]] == 'G'):
               env1[crdnt[0]][crdnt[1] + 1] = 'A'
               env1[crdnt[0]][crdnt[1]] = 'E'
               self.currentState = [crdnt[0], crdnt[1] + 1]
               return(env1, -100)
        else:
            return(env1, 0)
###################################################################################################
###################################################################################################
    def DownAction(self,env):
    
        crdnt = self.GetCurrentState(env)
        env1 = copy.deepcopy(env)
        if(crdnt[0] <= len(env) - 2):
            if(env1[crdnt[0] + 1][crdnt[1]] == 'D'):
               env1[crdnt[0] + 1][crdnt[1]] = 'A'
               env1[crdnt[0]][crdnt[1]] = 'E'
               self.currentState = [crdnt[0] + 1, crdnt[1]]
               return(env1, 1)
            if(env1[crdnt[0] + 1][crdnt[1]] == 'W'):
                return(env1, 0)
            if(env1[crdnt[0] + 1][crdnt[1]] == 'E'):
               env1[crdnt[0] + 1][crdnt[1]] = 'A'
               env1[crdnt[0]][crdnt[1]] = 'E'
               self.currentState = [crdnt[0] + 1, crdnt[1]]
               return(env1, 0)
            if(env1[crdnt[0] + 1][crdnt[1]] == 'G'):
               env1[crdnt[0] + 1][crdnt[1]] = 'A'
               env1[crdnt[0]][crdnt[1]] = 'E'
               self.currentState = [crdnt[0] + 1, crdnt[1]]
               return(env1, -100)
        else:
            return(env1, 0)
###################################################################################################
###################################################################################################
    def LeftAction(self, env):
        crdnt = self.GetCurrentState(env)
        env1 = copy.deepcopy(env)
        if(crdnt[1] >= 1):
            if(env1[crdnt[0]][crdnt[1] - 1] == 'D'):
               env1[crdnt[0]][crdnt[1] - 1] = 'A'
               env1[crdnt[0]][crdnt[1]] = 'E'
               self.currentState = [crdnt[0], crdnt[1] - 1]
               return(env1, 1)
            if(env1[crdnt[0]][crdnt[1] - 1] == 'W'):
                return(env1, 0)
            if(env1[crdnt[0]][crdnt[1] - 1] == 'E'):
               env1[crdnt[0]][crdnt[1] - 1] = 'A'
               env1[crdnt[0]][crdnt[1]] = 'E'
               self.currentState = [crdnt[0], crdnt[1] - 1]
               return(env1, 0)
            if(env1[crdnt[0]][crdnt[1] - 1] == 'G'):
               env1[crdnt[0]][crdnt[1] - 1] = 'A'
               env1[crdnt[0]][crdnt[1]] = 'E'
               self.currentState = [crdnt[0], crdnt[1] - 1]
               return(env1, -100)
        else:
            return(env1, 0)
###################################################################################################
###################################################################################################
    def GetCurrentState(self, env):
        for i in range(len(env)):
            for j in range(len(env[i])):
                if(env[i][j] == 'A'):
                    return([i, j])
###################################################################################################
    def Play(self, env):
        # 0:UP , 1:RIGHT , 2:DOWN , 3:LEFT
        action = self.TrueAction(env)
        i = 0
        while(self.Terminate(env) > 0 and self.currentState != [4, 4]) :
            action = self.TrueAction(env)
            if(action == 0):
                tup = self.UpAction(env)
                env1 = copy.deepcopy(tup[0])
                rew = tup[1]
                self.UpadteQvalue(env, env1, action, rew)
                env = copy.deepcopy(env1)
            elif(action == 1):
                tup = self.RightAction(env)
                env1 = copy.deepcopy(tup[0])
                rew = tup[1]
                self.UpadteQvalue(env, env1, action, rew)
                env = copy.deepcopy(env1)
            elif(action == 2):
                tup = self.DownAction(env)
                env1 = copy.deepcopy(tup[0])
                rew = tup[1]
                self.UpadteQvalue(env, env1, action, rew)
                env = copy.deepcopy(env1)
            elif(action == 3):
                tup = self.LeftAction(env)
                env1 = copy.deepcopy(tup[0])
                rew = tup[1]
                self.UpadteQvalue(env, env1, action, rew)
                env = copy.deepcopy(env1)


                
        if(self.currentState == [4, 4]):
            return("Loss")
        elif(self.Terminate(env) == 0):
            return("Win")
###################################################################################################
    def TrueAction(self, env):
        if(env not in states):
            states.append(env)
            statesIndx.append([0 ,0, 0, 0])
            action = random.randint(0 ,3)
            return(action)
        else:
            indx = states.index(env)
            envIndx = statesIndx[indx]
            mx = max(envIndx)
            l = []
            for i in range(4):
                if(envIndx[i] == mx):
                    l.append(i)

            if(len(l) >= 1):
                action = random.randint(0, len(l) - 1)
            else:
                action = l[0]
            return(l[action])
###################################################################################################
###################################################################################################
def runEpisodes(numEp, env, gama, alpha):
    for i in range(1, numEp + 1):
        Ag = Agent(gama, alpha)
        if(i % 100 == 0):
            print(f'game {i} passed')
            print(Ag.Play(env))

    print(len(states))
###################################################################################################

env = [ ['A', 'D', 'D', 'D', 'D', 'D', 'D', 'D', 'D']
       ,['D', 'W', 'W', 'W', 'D', 'W', 'W', 'W', 'D']
       ,['D', 'W', 'D', 'D', 'D', 'D', 'D', 'W', 'D']
       ,['D', 'D', 'D', 'W', 'E', 'W', 'D', 'D', 'D']
       ,['D', 'W', 'D', 'W', 'G', 'W', 'D', 'W', 'D']
       ,['D', 'W', 'D', 'D', 'W', 'D', 'D', 'W', 'D']
       ,['D', 'D', 'D', 'D', 'D', 'D', 'D', 'D', 'D'] ]

# env = []
# a, b = map(int,input('enter the dimensions').split())
# for i in range(a):
#     a = []
#     for j in range(b):
#         x = input(f'enter the state{i},{j}')
#         a.append(x)
#     env.append(a)

numEp = int(input('Number Of Episodes For Training : '))
gama = float(input('please input the gama value :'))
alpha = float(input('please input the alpha value :'))

runEpisodes(numEp,env, gama, alpha)
```
