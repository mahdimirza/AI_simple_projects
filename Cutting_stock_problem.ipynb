# Hill_Climbing

```python
import random
import copy
import math
```

```python


class Solution:
    def __init__(self, req, stock):
        self.solution = req                   # intialize by input list
        self.stock = stock                    # length of stock

#############################################################################
    def generate_solution(self):
        random.shuffle(self.solution)           # generating a random solution by shuffling list

##############################################################################

# this function calculate the number of stocks we need for this order of list

    def calc_cost(self, req):
        res = 0
        sum = 0
        for i in range(len(req)):
            if(sum + req[i] < self.stock) :
                sum += req[i]
            else:
                res += 1
                sum = req[i]

        return res

################################################################################

# this function Runs HillClimbing Algorithm on random generated solution

    def Hill_Climbing(self):
        res = self.calc_cost(self.solution)
        while True:
            flag = 0
            for i in range(len(self.solution) - 1):
                for j in range(i, len(self.solution)):
                    ls = self.solution.copy()
                    ls[i], ls[j] = ls[j], ls[i]
                    if(self.calc_cost(ls) < res):
                        self.solution = ls
                        res = self.calc_cost(ls)
                        flag = 1


            if(flag == 0):
                break

        return(res)

####################################################################################

#print best order of input list which has been found

    def get_solution(self):
        print(self.solution)

####################################################################################

mylist1 = [106, 187, 914, 106, 33, 18, 402, 230, 507, 495, 609, 627, 346, 295, 312, 107, 716, 88, 106, 248,
          689, 115, 106, 218, 672, 618, 117, 805, 306, 753, 414, 84, 557, 266, 409, 144, 69, 116, 333, 88,
          264, 967, 180, 251, 71, 788, 581, 555, 988, 292, 60, 125, 532, 405, 170, 249, 181, 686, 283, 424,
          933, 23, 99, 135, 246, 337, 648, 753, 354, 518, 45, 286, 315, 370, 557, 463, 312, 284, 61, 412, 457,
          118, 268, 123, 232, 788, 678, 371, 171, 557, 549, 286, 356, 92, 148, 515, 301, 632, 987, 660, 868,
          92, 544, 211, 70, 75, 145, 125, 278, 441, 368, 351, 119, 662, 653, 186, 517, 43, 224, 506, 592, 501,
          149, 79, 241, 53, 80, 437, 46, 78, 149, 525, 149, 126, 365, 460, 280, 266, 109, 86]

#answer = 51 or less than 51

S1 = Solution(mylist1, 1000)  # you must assign stock length here ##################################
S1.generate_solution()
res = S1.Hill_Climbing()
print(res)

S1.get_solution()
```
